```python
import numpy as np # linear algebra
import pandas as pd 

dataset = pd.read_csv("/Users/Helena/desktop/travel-insurance.csv")
```

```python
print(dataset.dtypes)
```

```python
dataset.dropna(how='any', inplace=True) 
dataset.drop(['Agency', 'Agency Type', 'Distribution Channel', 'Product Name', 'Claim', 'Duration', 'Destination', 'Gender'], axis = 1) 
```

```python
print(dataset["Net Sales"].mean())
print(dataset["Age"].mean())
```

```python
import matplotlib as plt
%matplotlib inline
```

```python
dataset["Net Sales"].plot.hist()
```

```python
dataset["Age"].plot.hist() 
```

```python
mean = (dataset.groupby(['Net Sales', 'Age']).mean())
print(mean)
```

```python
# get the maximum values of the Age column in dataframe
 
dataset['Age'].max()
```

```python
# get the minimum values of the Age column in dataframe

dataset['Age'].min()
```

### Conclusion

The maximum age that is willing to pay for the Travel Insurance is 118, and minimum age is 0 which I assume people would pay for the insurance when they are ready to travel, there is no necessary age for people to pay for the insurance. If 118 is the maximum age, then the Travel Insurance market is pretty big
